```python
#Carregando dados de ordem 
ordens = pd.read_csv("../data/raw/olist_orders_dataset.csv")
ordens.info()
```

```python
#Listando colunas
colunas_ordens_alterar = [
    'order_purchase_timestamp',
    'order_approved_at',
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date'
]
#Convertendo colunas de data/hora que estão como strings para datetime
ordens[colunas_ordens_alterar] = ordens[colunas_ordens_alterar].apply(pd.to_datetime)
```

```python
#Convertendo colunas de data/hora que estão como strings para datetime
order_items = pd.read_csv("../data/raw/olist_order_items_dataset.csv")
order_items.info()
```

```python
#Convertendo colunas de data/hora que estão como strings para datetime
order_items['shipping_limit_date'] = pd.to_datetime(order_items['shipping_limit_date'])
order_items.info()
```

```python
#Carregando dados
order_reviews = pd.read_csv("../data/raw/olist_order_reviews_dataset.csv")
```
