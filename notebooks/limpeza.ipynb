```python
#Importando bibliotecas
import pandas as pd 
```

```python
#Carregando dados de ordem 
ordens = pd.read_csv("../data/raw/olist_orders_dataset.csv")
ordens.info()
```

```python
#Listando colunas
colunas_ordens_alterar = [
    'order_purchase_timestamp',
    'order_approved_at',
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date'
]
#Convertendo colunas de data/hora que estão como strings para datetime
ordens[colunas_ordens_alterar] = ordens[colunas_ordens_alterar].apply(pd.to_datetime)
```

```python
#Convertendo colunas de data/hora que estão como strings para datetime
order_items = pd.read_csv("../data/raw/olist_order_items_dataset.csv")
order_items.info()
```

```python
#Convertendo colunas de data/hora que estão como strings para datetime
order_items['shipping_limit_date'] = pd.to_datetime(order_items['shipping_limit_date'])
order_items.info()
```

```python
#Carregando dados
order_reviews = pd.read_csv("../data/raw/olist_order_reviews_dataset.csv")
```

```python
#Listando colunas
reviews_date_cols = [
    'review_creation_date',
    'review_answer_timestamp'
]

#Convertendo colunas de data/hora que estão como strings para datetime
order_reviews[reviews_date_cols] = order_reviews[reviews_date_cols].apply(pd.to_datetime)
order_reviews.info()
```

```python
#carregando dados restantes para o merge
df_customers = pd.read_csv("../data/raw/olist_customers_dataset.csv")
df_products = pd.read_csv("../data/raw/olist_products_dataset.csv")
df_products = pd.read_csv("../data/raw/olist_products_dataset.csv")
df_sellers = pd.read_csv("../data/raw/olist_sellers_dataset.csv")
```

```python
df_products.info()
```

```python
#Realizando joins nas tabelas
df_merged = (
    ordens
    .merge(df_customers, on="customer_id", how="left")
    .merge(order_items, on="order_id", how="left")
    .merge(df_products, on="product_id", how="left")
    .merge(df_sellers, on="seller_id", how="left")
)
df_merged.head()
```

```python
df_merged.to_csv("../data/processed/olist_processado.csv", index=False)
```
